# Darbas su labai dideliu duomenų rinkiniu

```python
large_df = pd.DataFrame(columns=df.columns)

for temp_df in pd.read_csv('pokemon_data.csv', chunksize=5):
    results = temp_df.groupby(['Type 1']).count()
    
    large_df = pd.concat([large_df, results])
    print(large_df)
```
