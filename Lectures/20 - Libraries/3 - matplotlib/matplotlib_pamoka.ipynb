```python
import matplotlib.pyplot as plt
import numpy as np
```

# Paprasta diagrama

## Sukuriam linijinę diagramą

```python
plt.plot([1, 2, 3], [2, 4, 6])
```

## Linijinei diagramai duodam duomenis iš sąrašų

```python
x = [1, 2, 3]
y = [2, 4, 6]

plt.plot(x, y)
```

## Diagramos piešimą iškviečiam kitaip

```python
x = [1, 2, 3]
y = [2, 4, 6]

plt.plot(x, y)

plt.show()
```

## Diagramai pridedam tekstus

```python
x = [1, 2, 3]
y = [2, 4, 6]

plt.plot(x, y)

plt.title('Pirmas grafikas!')

plt.xlabel('X ašis')
plt.ylabel('Y ašis')

plt.show()
```

## Atnaujinam diagramos tekstų parametrus

```python
x = [1, 2, 3]
y = [2, 4, 6]

plt.plot(x, y)

plt.title('Pirmas grafikas!', fontdict={'fontname': 'Open sans', 'fontsize': 20})

plt.xlabel('X ašis', fontdict={'fontsize': 12})
plt.ylabel('Y ašis', fontdict={'fontsize': 12})

plt.show()
```

## Sukuriam savo ašis

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

plt.plot(x, y)

plt.title('Pirmas grafikas!', fontdict={'fontname': 'Open sans', 'fontsize': 20})

plt.xlabel('X ašis', fontdict={'fontsize': 12})
plt.ylabel('Y ašis', fontdict={'fontsize': 12})

plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 2, 4, 6, 8, 10])

plt.show()
```

## Diagrama prisitaiko kad ir kokios ašys būtų sudėtos

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

plt.plot(x, y)

plt.title('Pirmas grafikas!', fontdict={'fontname': 'Open sans', 'fontsize': 20})

plt.xlabel('X ašis', fontdict={'fontsize': 12})
plt.ylabel('Y ašis', fontdict={'fontsize': 12})

plt.xticks([0, 1, 2, 3, 4, 10])
plt.yticks([0, 2, 4, 6, 7, 7.5, 8, 10])

plt.show()
```

## Legend pridėjimas

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# čia uždedam kaip vadinam šitą liniją
plt.plot(x, y, label='2x')

plt.title('Pirmas grafikas!', fontdict={'fontname': 'Open sans', 'fontsize': 20})

plt.xlabel('X ašis', fontdict={'fontsize': 12})
plt.ylabel('Y ašis', fontdict={'fontsize': 12})

plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 2, 4, 6, 8, 10])

# iškviečiam legend rodymą
plt.legend()

plt.show()
```

## Formatuojam liniją

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# pokyčiai šioje eilutėje
plt.plot(x, y, label='2x', color='red', linewidth=2, linestyle='--', marker='.', markersize=10, markeredgecolor='blue')
# plt.plot(x, y, label='2x', color='yellow')
# plt.plot(x, y, label='2x', color='y')
# plt.plot(x, y, label='2x', color='#ababab')

plt.title('Pirmas grafikas!', fontdict={'fontname': 'Open sans', 'fontsize': 20})

plt.xlabel('X ašis', fontdict={'fontsize': 12})
plt.ylabel('Y ašis', fontdict={'fontsize': 12})

plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()

plt.show()
```

## Formatavimo santrumpa

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# plt.plot(x, y, label='2x', color='red', linewidth=2, linestyle='--', marker='.', markersize=10, markeredgecolor='blue')

# Sutrumpinimo santrumpa
# fmt = '[color][marker][line]'
# plt.plot(x, y, 'r.-', label='2x')
# plt.plot(x, y, 'r.--', label='2x')
# plt.plot(x, y, 'r^--', label='2x')
plt.plot(x, y, 'b^--', label='2x')

plt.title('Pirmas grafikas!', fontdict={'fontname': 'Open sans', 'fontsize': 20})

plt.xlabel('X ašis', fontdict={'fontsize': 12})
plt.ylabel('Y ašis', fontdict={'fontsize': 12})

plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()

plt.show()
```

## Pridedam antrą liniją

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# Pirma linija

plt.plot(x, y, 'b^--', label='2x')

# Antra linija

x2 = np.arange(0, 4.5, 0.5)
# print(x2)
plt.plot(x2, x2 ** 2, 'red', label='x^2') # x ašis iš duomenų, y ašis x'o kvadratai

# Kiti parametrai

plt.title('Pirmas grafikas!', fontdict={'fontname': 'Open sans', 'fontsize': 20})

plt.xlabel('X ašis', fontdict={'fontsize': 12})
plt.ylabel('Y ašis', fontdict={'fontsize': 12})

plt.xticks([0, 1, 2, 3, 4])
# plt.yticks([0, 2, 4, 6, 8, 10]) # užkomentuojam šitą

plt.legend()

plt.show()
```

## Tai pačiai linijai taikome skirtingus stilius

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# Pirma linija

plt.plot(x, y, 'b^--', label='2x')

# Antra linija

x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2[:6], x2[:6] ** 2, 'r', label='x^2') # spausdinam pirmus 6 taškus
plt.plot(x2[5:], x2[5:] ** 2, 'r--') # spausdinam likusius taškus

# Kiti parametrai

plt.title('Pirmas grafikas!', fontdict={'fontname': 'Open sans', 'fontsize': 20})

plt.xlabel('X ašis', fontdict={'fontsize': 12})
plt.ylabel('Y ašis', fontdict={'fontsize': 12})

plt.xticks([0, 1, 2, 3, 4])
# plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()

plt.show()
```

## Grafiko dydžio pakeitimas

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# Grafiko dydžio pakeitimas

# plt.figure(figsize=(5, 3)) # 5, 3 - aspect ratio
# plt.figure(figsize=(5, 3), dpi=300) # dpi - pixels per inch, daugiau yra kokybiškiau
plt.figure(figsize=(5, 3), dpi=200) # dpi - pixels per inch, daugiau yra kokybiškiau

# Pirma linija

plt.plot(x, y, 'b^--', label='2x')

# Antra linija

x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2[:6], x2[:6] ** 2, 'r', label='x^2')
plt.plot(x2[5:], x2[5:] ** 2, 'r--')

# Kiti parametrai

plt.title('Pirmas grafikas!', fontdict={'fontname': 'Open sans', 'fontsize': 20})

plt.xlabel('X ašis', fontdict={'fontsize': 12})
plt.ylabel('Y ašis', fontdict={'fontsize': 12})

plt.xticks([0, 1, 2, 3, 4])
# plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()

plt.show()
```

## Grafiko išsaugojimas faile

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# Grafiko dydžio pakeitimas

plt.figure(figsize=(5, 3), dpi=200)

# Pirma linija

plt.plot(x, y, 'b^--', label='2x')

# Antra linija

x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2[:6], x2[:6] ** 2, 'r', label='x^2')
plt.plot(x2[5:], x2[5:] ** 2, 'r--')

# Kiti parametrai

plt.title('Pirmas grafikas!', fontdict={'fontname': 'Open sans', 'fontsize': 20})

plt.xlabel('X ašis', fontdict={'fontsize': 12})
plt.ylabel('Y ašis', fontdict={'fontsize': 12})

plt.xticks([0, 1, 2, 3, 4])
# plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()

# Grafiko išsaugojimas faile

# plt.savefig('pirmas-grafikas.png')
plt.savefig('pirmas-grafikas.png', dpi=300)

plt.show()
```

# Antras grafikas

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

plt.bar(labels, values)

plt.figure(figsize=(6, 4))

plt.show()
```

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values)

bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('*')

plt.figure(figsize=(6, 4))

plt.show()
```

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values)

patterns = ['/', 'O', '*']

for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.figure(figsize=(6, 4))

plt.show()
```
