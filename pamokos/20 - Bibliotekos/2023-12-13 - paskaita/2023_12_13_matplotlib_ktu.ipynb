```python
metai = [2000, 2001, 2002, 2003, 2004]
pardavimai_mil = [3.5, 3.6, 3.2, 3.3, 3.7]

plt.plot(metai, pardavimai_mil)

plt.title('Pardavimai pagal metus')

plt.xlabel('Metai')
plt.ylabel('Pajamos, mil.')

# plt.xticks([2000, 2005, 2010])
plt.xticks(metai)
# plt.yticks([])

plt.show()
```

```python
metai = [2019, 2020, 2021, 2022]
ktu_studentu_kiekis = [20_000, 21_000, 20_500, 23_000]
vdu_studentu_kiekis = [21_350, 22_350, 23_035, 21_356]
ku_studentu_kiekis = [10478, 12034, 11003, 13057]

plt.plot(metai, ktu_studentu_kiekis, color='red', linestyle='--', marker='o')
plt.plot(metai, vdu_studentu_kiekis, color='blue')
plt.plot(metai, ku_studentu_kiekis, color='#B185A7')

plt.xticks(metai)

plt.show()
```
